```python
##########################################################################
##           DATA SCIENCE: S03.T01: Nivell 1 Exercici 1                 ##
##########################################################################
# Exercici 1: Resumeix gràficament el data set DelayedFlights.csv
# Crea almenys una visualització per:

#     1.- Una variable categòrica (UniqueCarrier)
#     2.- Una variable numèrica (ArrDelay)
#     3.- Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)
#     4.- Dues variables numèriques (ArrDelay i DepDelay)
#     5.- Tres variables (ArrDelay, DepDelay i UniqueCarrier)
#     6.- Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier).

#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


#************************************************************************************
#********************************         Main       ********************************
#************************************************************************************
#Read *.csv file (Import Data)
df = pd.read_csv("Data sets\DelayedFlights.csv", usecols=("DepTime", 
                                                          "CRSDepTime",
                                                          "ArrTime",
                                                          "UniqueCarrier",
                                                          "FlightNum",
                                                          "TailNum",
                                                          "ActualElapsedTime",
                                                          "CRSElapsedTime",
                                                          "AirTime",
                                                          "ArrDelay",
                                                          "DepDelay",
                                                          "Origin",
                                                          "Dest",
                                                          "Distance",
                                                          "CarrierDelay",
                                                          "NASDelay",
                                                          "WeatherDelay",
                                                          "SecurityDelay",
                                                          "LateAircraftDelay"), )#nrows=10000)

print ("Versió de Panda: ", pd.__version__)                  #Print version of PANDA
display(df)                                                  #Print Dataframe
```

```python
#************************************************************************************
#********************************      Functions     ********************************
#************************************************************************************

# Save a Plot how an Image
def saveImage(figure,name):
    figure.savefig(name)

#Function plot, type Histogram (one categoric value)
def plotHistogram(df):
    
    # create a dataset
    height = df
    bars = df.index
    x_pos = np.arange(len(bars))

    # Create bars with different colors
    #plt.bar(x_pos, height, color=['black', 'red', 'green', 'blue', 'cyan'])
    plt.bar(x_pos, height, color=['green'])
    # Create names on the x-axis
    plt.xticks(x_pos,bars)

    # Show graphic
    fig = plt.gcf()
    plt.show()

    return fig

#Function plot, type Histogram (one Numeric value)
def plotHistogram2(df,paramNum):
    
    # create a dataset
    height = df[paramNum]
    bars = df.index
    x_pos = np.arange(len(bars))

    # Create bars with different colors
    #plt.bar(x_pos, height, color=['black', 'red', 'green', 'blue', 'cyan'])
    plt.bar(x_pos, height, color=['yellow'])
    # Create names on the x-axis
    plt.xticks(x_pos,bars)

    # Show graphic
    fig = plt.gcf()
    plt.show()

    return fig

#Function plot, type Density Chart (one numeric value)
def plotDensityChart(df, paramNum):
    # library & dataset
    import seaborn as sns
    
    # set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
    sns.set(style="darkgrid")
    
    # Plot the histogram thanks to the distplot function
    sns.kdeplot(df[paramNum])

    # Show graphic
    fig = plt.gcf()
    plt.show()

    return fig    
    
#Function plot, type Box Plot (with 1 numeric value and 1 categoric value)   
def plotBoxPlot(df, paramCat, paramNum):
    # library & dataset
    import seaborn as sns
    
    # set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
    sns.set(style="darkgrid")
    
    # Plot the histogram thanks to the distplot function
    sns.boxplot( x=df[paramCat], y=df[paramNum], color='skyblue' )

    # Show graphic
    fig = plt.gcf()
    plt.show()

    return fig    
    
#Function plot, type Box Plot (with 2 numeric values)   
def plotDensityPlot(df, paramNum1, paramNum2):
    # library & dataset
    import seaborn as sns
    
    # set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
    sns.set(style="darkgrid")

    # plotting both distibutions on the same figure
    fig = sns.kdeplot(df[paramNum1], shade=True, color="r")
    fig = sns.kdeplot(df[paramNum2], shade=True, color="b")

    # Show graphic
    fig = plt.gcf()
    plt.show()

    return fig

#Function plot, type Stacked Area Chart (with 3 numeric values and 1 categoric value)   
def plotBarplotMultiple(df, paramCat1, paramNum1, paramNum2, paramNum3, Sel3):
        # libraries
        import numpy as np
        import matplotlib.pyplot as plt

        # width of the bars
        barWidth = 0.3
        
        # Choose the height of the blue bars
        bars1 = df[paramNum1] 

        # Choose the height of the cyan bars
        bars2 = df[paramNum2] 

        if (Sel3 == True):
            # Choose the height of the cyan bars
            bars3 = df[paramNum3] 
        
        # Choose the height of the error bars (bars1)
        yer1 = np.arange(len(bars1))

        # Choose the height of the error bars (bars2)
        yer2 = np.arange(len(bars1))
        
        #print(title) by groups of bars
        title = df.index 
        
        # The x position of bars
        r1 = np.arange(len(bars1))
        r2 = [x + barWidth for x in r1]
        r3 = [x + barWidth for x in r2]

        # Create blue bars
        plt.bar(r1, bars1, width = barWidth, color = 'blue', edgecolor = 'black', yerr=yer1, capsize=7, label=paramNum1)

        # Create cyan bars
        plt.bar(r2, bars2, width = barWidth, color = 'yellow', edgecolor = 'black', yerr=yer2, capsize=7, label=paramNum2)

        if (Sel3 == True):
             # Create cyan bars
             plt.bar(r3, bars3, width = barWidth, color = 'green', edgecolor = 'black', yerr=yer2, capsize=7, label=paramNum3)
   
        # general layout
        plt.xticks([r + barWidth for r in range(len(bars1))], title)
        plt.ylabel('height')
        plt.legend()

        # Show graphic
        fig = plt.gcf()
        plt.show()

        return fig

#Function plot, type Box Plot (with 2 numeric values)   
def plotArea(df, paramNum1, paramNum2):
    # library & dataset
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots()
    
    #Define Axis X and Y
    axisX = df.index
    axisY = df[paramNum1]
    
    #Set labels
    ax.set_title('Gràfica de: ' + paramNum1)
    ax.set_xlabel('Eix X: '+ paramNum2)
    ax.set_ylabel('Eix Y: '+ paramNum1)

    #Show axis grid
    ax.xaxis.grid(True)
    ax.yaxis.grid(True)
    #Axis Grid Style
    ax.xaxis.grid(color='r', linestyle='-')  # equivale a color='red',  linestyle='solid'
    ax.yaxis.grid(color='b', linestyle='-')  # equivale a color='blue', linestyle='solid'
    
    ax.plot(axisX, axisY)
    plt.show()
    #ax.fill_between(axisX, axisY)
    #plt.show()

    # Show graphic
    #fig = plt.gcf()
    plt.show()

    return fig
    
#************************************************************************************
#********************************         Main       ********************************
#************************************************************************************
# Create a dataframe group by a categoric value
dfAirlines = df.groupby(["UniqueCarrier"]).sum().sort_values(by="UniqueCarrier", ascending=True)
display(dfAirlines)

# (1) Create a dataframe by calculate Result (1)
df3 = df['UniqueCarrier'].value_counts() 
display(df3)

# Print Result (1)
print ("(1) Gràfica amb una variable categòrica (UniqueCarrier):")          #Print message
fig1 = plotHistogram(df3)                                                   #Call Function create plot
saveImage(fig1,'Images/Image_Result_1.png')                                        #Save Image

#Print Result (2)
print ("(2) Gràfica amb una variable Numèrica (ArrDelay):")                 #Print message
fig1 = plotDensityChart(df,"ArrDelay")                                      #Call Function create plot type Density Chart
saveImage(fig1,'Images/Image_Result_2.png')                                 #Save Image

#Print Result (3)
print ("(3) Gràfica amb una variable numèrica i una categòrica (ArrDelay i UniqueCarrier):")    #Print message
fig1 = plotBoxPlot(df,"UniqueCarrier","ArrDelay")                           #Call Function create plot type Density Chart
saveImage(fig1,'Images/Image_Result_3.png')                                 #Save Image

#Print Result (4)
print ("(4) Gràfica amb dues variables numèriques (ArrDelay i DepDelay):")  #Print message
fig1 = plotDensityPlot(df,"DepDelay","ArrDelay")                            #Call Function create plot type Density Chart
saveImage(fig1,'Images/Image_Result_4.png')                                 #Save Image



```

```python
#************************************************************************************
#********************************         Main       ********************************
#************************************************************************************
#Print Result (5)
print ("(5) Gràfica amb tres variables (ArrDelay, DepDelay i UniqueCarrier):")   #Print message
fig1 = plotBarplotMultiple(dfAirlines,"UniqueCarrier","DepDelay","ArrDelay","AirTime",False)               #Call Function create plot type Density Chart
saveImage(fig1,'Images/Image_Result_5.png')

#Print Result (6)
print ("(6) Gràfica amb Més tres variables numèriques (ArrDelay, DepDelay, AirTime i UniqueCarrier):")  
fig1 = plotBarplotMultiple(dfAirlines,"UniqueCarrier","DepDelay","ArrDelay","AirTime",True)               #Call Function create plot type Density Chart
saveImage(fig1,'Images/Image_Result_6.png')

```
